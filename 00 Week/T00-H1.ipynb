```python

import pandas as pd
df = pd.read_csv('LaborSupply1988.csv')
print(df.to_string())

print('Anzahl Spalten: ') 
print(len(df.columns))

print('Anzahl Zeilen: ') 
print(len(df))


```

```python
print("Die ersten 10 Data sets sind: ")
print(df.head(10))

min_age = df['age'].min()
max_age = df['age'].max()
range = max_age - min_age
print("Die range ist: ",range)

```

```python
average_lnhr_by_kids = df.groupby('kids')['lnhr'].mean()
print(average_lnhr_by_kids)

averageNrKidsOfFourtyYearsOld = df.groupby('age')['kids'].mean()
print ('40 Jährige haben im Durchschnitt:',round(averageNrKidsOfFourtyYearsOld[40],2), ' Kinder')
```

### Aufgabe 2


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the histogram of the 'age' attribute
plt.figure(figsize=(10, 6))  # Set the figure size for better visualization
sns.histplot(df['age'], bins=20, kde=False, color='skyblue')  # Plot a histogram, no KDE curve
plt.title('Histogram of Age')  # Set the title of the plot
plt.xlabel('Age')  # Label the x-axis
plt.ylabel('Frequency')  # Label the y-axis
plt.grid(True)  # Add gridlines for clarity
plt.show()
```

```python
avg_Kids_Age = df.groupby('age')['kids'].mean()

#Line Plot Average Numver of kids by Age
plt.figure(figsize=(12, 6))  # Set the figure size for better visualization
sns.lineplot(x= avg_Kids_Age.index, y = avg_Kids_Age.values, color='skyblue', marker="o")  # Plot a histogram2
plt.title('Average number of kids by Age')  # Set the title of the plot
plt.xlabel('Age')  # Label the x-axis
plt.ylabel('Number of kids')  # Label the y-axis
plt.grid(True)  # Add gridlines for clarity
plt.show()

#Correlation
correlation = df['kids'].corr(df['age'])
print ("Die Korrelation zwischen Alter und Kinder ist:", round(correlation,2))
```

```python
avg_Kids_Age = df.groupby('age')['lnwg'].mean()

# Scatter Plot of 'lnwg' against 'age'
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='lnwg', data=df, alpha=0.5)
plt.title('Log of Hourly Wage (lnwg) vs Age')
plt.xlabel('Age')
plt.ylabel('Log of Hourly Wage (lnwg)')
plt.grid(True)
plt.show()

#Line Plot log of hourly wage against age
plt.figure(figsize=(12, 6))  # Set the figure size for better visualization
sns.lineplot(x= avg_Kids_Age.index, y = avg_Kids_Age.values, color='skyblue', marker="o")  # Plot a line diagram
plt.title('log of hourly average wage against age')  # Set the title of the plot
plt.xlabel('Age')  # Label the x-axis
plt.ylabel('Log of the hourly wage')  # Label the y-axis
plt.grid(True)  # Add gridlines for clarity
plt.show()

correlation = df['lnwg'].corr(df['age'])
print(f"The correlation between 'lnwg' and 'age' is: {correlation:.2f}")
```

```python
# Scatter plot of 'lnhr' vs 'age' with different colors for 'disab' status
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='lnhr', hue='disab', data=df, palette={0: "blue", 1: "red"}, alpha=0.6)
plt.title('Log of Annual Hours Worked (lnhr) vs Age by Disability Status')
plt.xlabel('Age')
plt.ylabel('Log of Annual Hours Worked (lnhr)')
plt.legend(title='Disability Status', labels=['No Disability (0)', 'Disability (1)'])
plt.grid(True)
plt.show()
```

```python
# Boxplot of 'lnhr' against 'kids'
plt.figure(figsize=(12, 6))
sns.boxplot(x='kids', y='lnhr', data=df, palette="coolwarm")
plt.title('Boxplot of Log of Annual Hours Worked (lnhr) by Number of Kids')
plt.xlabel('Number of Kids')
plt.ylabel('Log of Annual Hours Worked (lnhr)')
plt.grid(True)
plt.show()
```
