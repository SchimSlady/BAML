e) What is the value range of the attribute "Beer consumption (million liters)"?

```python
min_Beer = visitors['Beer consumption (million liters)'].min()
max_Beer = visitors['Beer consumption (million liters)'].max()
range = max_Beer - min_Beer
print ("Die range ist:", round (range, 2))
```

f) What was the year with the greatest beer consumption? How much beer was consumed?

```python
max_beer_year = visitors.loc[visitors['Beer consumption (million liters)'].idxmax()]

max_year = max_beer_year['Year']
max_beer = max_beer_year['Beer consumption (million liters)']

print("In", round(max_year), max_beer, "Liter Bier")
```

g) Plot the beer consumption over the years. Turn on the grid of your plot.

```python
#Line Plot beer consumption over the years
plt.figure(figsize=(12, 6))  # Set the figure size for better visualization
sns.lineplot(x= visitors['Year'], y = visitors['Beer consumption (million liters)'], color='skyblue', marker="o")  # Plot 
plt.title('beer consumption over the years')  # Set the title of the plot
plt.xlabel('Year')  # Label the x-axis
plt.ylabel('Beer consumption (million liters)')  # Label the y-axis
plt.grid(True)  # Add gridlines for clarity
plt.show()
```

i) Load the data from *beer_price.csv* into a second dataframe. Then merge the two dataframes based on the year with an outer join.  
Observe which values you get for years that are not listed in some of the data sets.

```python
beerPrice = pd.read_csv('beer_price.csv')
merged_df = pd.merge(visitors, beerPrice, how='outer', on='Year')

merged_df
```

j) Until 2001, the beer price is reported in DEM ("Deutsche Mark"), the former German currency. 
Adjust the beer prices (min and max) to give values in EUR.
You may assume that 1 EUR = 1.95583 DEM. 

```python
conversion_rate = 1.95583
merged_df.loc[merged_df['Year'] <= 2001, 'Min price'] = (merged_df.loc[merged_df['Year']<= 2001, 'Min price'] / conversion_rate)
merged_df.loc[merged_df['Year'] <= 2001, 'Max price'] = (merged_df.loc[merged_df['Year']<= 2001, 'Max price'] / conversion_rate)

merged_df
```

k) Compute the mean and variance of "Min price" during the period from 2000 to 2007.

```python
mask = (2000 <= merged_df["Year"]) & (merged_df["Year"] <= 2007)
mean = merged_df[mask]["Min price"].mean()
var = merged_df[mask]["Min price"].var()
print("mean =", mean)
print("var =", var)
```

l) Add a new column to the merged dataframe, describing the relative difference in beer prices
between two consecutive years in percent. Plot this difference against "Year".

```python
merged_df ['Price increase (%)'] = (
    (merged_df['Min price']- merged_df["Min price"].shift(1))
    / merged_df["Min price"]
    * 100
)

merged_df.plot(x="Year", y='Price increase (%)')
```

m) Create a bar plot displaying the minimum and maximum prices over the years.

```python
merged_df.plot.bar(x="Year", y=["Min price", "Max price"])
```

n) Compute estimates (lower and upper bound) of the beer revenue and visualize them by bar plot.

```python
merged_df["Min sales"] = (
    merged_df["Beer consumption (million liters)"] * merged_df["Min price"]
)
merged_df["Max sales"] = (
    merged_df["Beer consumption (million liters)"] * merged_df["Max price"]
)

merged_df.plot.bar(x="Year", y=["Min sales", "Max sales"])
plt.title("Revenue")
plt.ylabel("Million EUR")
```
